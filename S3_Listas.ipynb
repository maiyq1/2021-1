```python
%%writefile listas.cpp

//PARTICIPARON TODOS

#include <iostream>
using namespace std;

struct Node{
    int elemento;
    Node* sig;
    Node(int e){ elemento = e; sig=nullptr; }
};

class Lista{
    Node* inicio;
    Node* final;
    size_t cantidad;
    public:
    Lista(){
        inicio = final = nullptr;
        cantidad = 0;
    }

    //Imprimir
    void print(){ 
        if(inicio == nullptr)
          cout << "La lista esta vacia" << endl;
        else{
            Node* aux = inicio;
            while(aux != nullptr){
                cout << aux->elemento << " ";
                aux = aux->sig;
            }
        }
    }

    //insertar al final O(n)
    void push_back(int v){
        Node* n = new Node(v);
        if(inicio == nullptr) //cantidad = 0
          inicio = n;
        else{
            Node* aux = inicio;
            while(aux->sig!=nullptr){ //recorrer hasta que el sig sea nullptr
                aux = aux->sig;
            }
            aux->sig = n;
        }
        ++cantidad;
    }

    //insertar al final O(1)
    void push_back_opt(int v){
        Node* n = new Node(v);
        if(inicio == nullptr)
          inicio = final = n;
        else{
            final->sig=n;
            final = final->sig;
        }
        ++cantidad;
    }

    //Insertar al inicio = O(1)
    void push_front(int v){
      Node* n = new Node(v);
      if(inicio == nullptr)
	      inicio = n;
      else{
	      n->sig=inicio;
	      inicio=n; 
      }  
      ++cantidad;
    }

    //Obtener objeto en posicion n
    void pick_node_pos(int pos){
        if(pos < 1 || pos > cantidad)
          cout << "Fuera de rango" << endl;
        else{
            Node* aux = inicio;
            size_t cont = 0;
            while(cont != pos){
                aux = aux->sig;
                ++cont;
            }
            cout << "El nodo de pos " << pos << " es " << aux->elemento;
        }
        cout << "\n";
    }

    //Buscar pos de nodo por valor
    void search_pos_by_value(int value){
        Node* aux = inicio;
        size_t cont;
        while(value != aux->elemento){
            aux = aux->sig;
            cont++;
        }
        cout << "El nodo de valor " << aux->elemento << " esta en la pos " << cont;
        cout << "\n";
    }
    //Ordenar lista de menor a mayor Nicola
        //Ordenar por valor
    void ordenar()
{   int op;
    cout<<"1. Ascendente \n 2. Descendente \n";
    cin >> op;
    Node* anterior = inicio;
    Node* actual = inicio->sig;
 if(op == 1){
    while (actual != NULL)
    {
        if (actual->elemento < anterior->elemento)
        {
            anterior->sig = actual->sig;
            actual->sig = inicio;
            inicio = actual;
            actual = anterior;
        }
        else
            anterior = actual;
        actual = actual->sig;
    }
 }
 else 
    while (actual != NULL)
      {
          if (actual->elemento > anterior->elemento)
          {
              anterior->sig = actual->sig;
              actual->sig = inicio;
              inicio = actual;
              actual = anterior;
          }
          else
              anterior = actual;
          actual = actual->sig;
      }
}
//buscar solo numeros pares Nicola
void buscarpares(){
        Node* aux = inicio;
        Node* actual = inicio->sig;
        while (actual != NULL){
            if (actual->elemento % 2 == 0)
        {
            cout << actual->elemento<<" ";
        }
        actual = actual->sig;
    }
    }
//Eliminar en posicion(n)
void eliminarElemento(size_t pos){
	Node* aux=inicio;
	Node* aux_eliminar=inicio;
	size_t cont=0;
	if(inicio!=NULL){
		while((aux_eliminar!=NULL) && (pos!=cont)){
			aux=aux_eliminar;
			aux_eliminar=aux_eliminar->sig;
			cont++;
		}
		if(aux_eliminar==NULL)
		{
			cout<<"El elemento de la no existe\n";
		}
		else if(aux_eliminar==inicio){
			inicio=inicio->sig;
			delete aux;
		}
		else{
			aux->sig=aux_eliminar->sig;
			delete aux_eliminar;
		}
	}
  else
  {
      cout<<"La lista esta vacia";
  }
}

//Eliminar al inicio = O(1)
    void delete_front(){
      Node* aux = inicio;
      Node* aux_eliminar=inicio;
      if(inicio != NULL)
      {
	      aux_eliminar=inicio;
	      aux=aux->sig;
		    inicio=aux;
		    delete aux_eliminar; 
      }  
      --cantidad;
    }
  
//Eliminar al final = O(n)
    void delete_back(){
      Node* aux = inicio;
      Node* aux_eliminar=inicio;
      if(inicio!=NULL){
      	if(aux->sig!=NULL){
	        while(aux_eliminar->sig!=NULL){
	      	  aux=aux_eliminar;
			      aux_eliminar=aux->sig;
		      }
		      aux->sig=aux_eliminar->sig;
		      delete aux_eliminar;
		      cantidad--;
	 	    }
	 	    else
	 	    {
	 	      delete aux_eliminar;
	 	      inicio=NULL;
		    }
     }
}

    //Asignacion
    //1.obtener objeto en posicion = at() OK
    //2.eliminar en posicion(n) OK Rafael
    //3.eliminar al inicio OK Rafael
    //4.eliminar al final OK Rafael
    //5.ordenar elementos(n^2)  OK Nicola
    //6.buscar por valor(n) OK
    //7.invertir lista(n^2)
    //8.buscar solo elementos pares OK Nicola
    //9.listar con uso de lambdas
    //10.listar de forma recursiva
};

int main(){
    Lista* miLista = new Lista();
    int value;
    int opc;
    while(opc != 10){
        cout << "\nTema: Listas enlazadas por nodos" << endl;
        cout << "Ingrese una opcion:" << endl;
        cout << "1. Imprimir lista\n2. Insertar al final O(n)\n";
        cout << "3. Insertar al inicio\n4. Obtener nodo de una posicion\n";
        cout << "5. Buscar nodo segun el valor\n6. Ordenar de menor a mayor\n";
        cout << "7. Eliminar en posicion\n8. Eliminar al inicio\n";
        cout << "9. Eliminar al final\n";
        cout << "Salir\n";
        cin >> opc;
        switch(opc){
            case 1:
              miLista->print();
              break;
            case 2:
              cout << "Ingrese el nuevo nodo: ";
              cin >> value;
              miLista->push_back(value);
              break;
            case 3:
              cout << "Ingrese el nuevo nodo inicial: ";
              cin >> value;
              miLista->push_front(value);
              break;
            case 4:
              cout << "Ingrese la posicion: ";
              cin >> value;
              miLista->pick_node_pos(value);
              break;
            case 5:
              cout << "Ingrese el valor del nodo: ";
              cin >> value;
              miLista->search_pos_by_value(value);
              break;
            case 6:
              miLista->ordenar();
              break;
            case 7:
              cout << "Ingrese la posicion: ";
              cin >> value;
              miLista->eliminarElemento(value);
              break;
            case 8:
              miLista->delete_front();
              break;
            case 9:
              miLista->delete_back();
              break;
        }
      }
    return 0;
}
```
